```python
from keras.models import Model
from keras.layers import Dense, Input
from keras.datasets import mnist 
from keras.utils import np_utils
import numpy as np
```

```python
#parametros comunes 
batch_size = 128
num_classes = 10
epoch = 100
learning_rate = 0.1
# MNISt
img_rows, img_cols = 28,28
nn_vissble = img_rows * img_cols
nn_hidden = 500
corruption_level = 0.3
```

```python
#Load MNIST Dataset
(x_train, _), (x_test, _) = mnist.load_data()
print('dataset original', x_train.shape)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
x_train = x_train.reshape((len(x_train),np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test),np.prod(x_test.shape[1:])))
```

```python
#AÃ±adiendo ruido a las imagenes 
x_train_noisy = x_train + corruption_level * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape)
x_test_noisy = x_test + corruption_level * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape)
x_train_noisy = np.clip(x_train_noisy,0.,1.)
x_test_noisy = np.clip(x_test_noisy,0.,1.)
```

```python
#Construimos el autoencoder
input_img = Input(shape=(nn_vissble,))
encoded = Dense(nn_hidden, activation='relu')(input_img)
decoded = Dense(nn_vissble, activation='sigmoid')(encoded) 

autoencoder = Model(input=input_img, output=decoded)
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
autoencoder.summary()
```

```python
#Entrenar 
autoencoder.fit(x_train_noisy, x_train, epochs=epoch,
               batch_size=batch_size, verbose=1,
               validation_data=(x_test_noisy, x_test))
```
