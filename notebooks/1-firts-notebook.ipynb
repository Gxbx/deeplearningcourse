```python
import numpy as np
import matplotlib as mp
import pandas as pd
from matplotlib import style
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("../data/train.csv")
```

```python
dataset.head()
```

```python
dataset_feature = dataset.drop(["PassengerId","Name","Embarked","Cabin","Parch"], axis=1)
```

```python
dataset_feature.tail()
```

```python
print("Sexo:", dataset_feature["Sex"].unique())
```

```python
dataset_numeric = pd.get_dummies(dataset_feature)
dataset_numeric.head()
```

```python
dataset_numeric.describe()
```

```python
#dataset_clean = dataset_numeric.copy()
#dataset_clean["Age"] = dataset_numeric.dropna(axis=1)
dataset_numeric["Age"].fillna(30)
```

```python
def plot_with_bar(values=None, title = None, xticks = None):
    if(values == None):
        raise ValueError("Values no puede ser vacio")
    if(type(values)==list) or (type(values)==tuple):
        pass
    else:
        raise TypeError("Values debe ser una lista, table o tupla")
    ind = [i for i in range(len(values))]
    plt.bar(ind, values)
    plt.title(title)
    plt.xticks(ind, xticks)
    plt.show()
```

```python
male = dataset_numeric.query("Sex_male == 1")
female = dataset_numeric.query("Sex_female == 1")
plot_with_bar (values = [male.shape[0], female.shape[0]], title = "Pasajeros por sexo", xticks = ["Hombre", "Mujeres"])
```

```python
female_survived = dataset_numeric.query("Survived == 1 and Sex_female == 1")
male_survived = dataset_numeric.query("Survived == 1 and Sex_male == 1")
plot_with_bar (values = [female_survived.shape[0], male_survived.shape[0]])
```
